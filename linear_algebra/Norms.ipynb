# Norms

Norms are used to measure the **magnitude** of a vector.

```python
class Norms:
    def l1_norm(self, x):
        return sum(abs(i) for i in x)

    def l2_norm(self, x):
        return sum(i**2 for i in x) ** (1/2)

    def squared_l2_norm(self, x):
        return sum(i**2 for i in x)

    def max_norm(self, x):
        return max(abs(i) for i in x)
    
    def generalized_lp_norm(self, x, p):
        return sum(abs(i) ** p for i in x) ** (1/p)
```

```python
n = Norms()
arr = [1,2,3,4,5]
```

```python
n.l1_norm(arr)
```

```python
n.l2_norm(arr)
```

```python
n.squared_l2_norm(arr)
```

```python
n.max_norm(arr)
```

```python
n.generalized_lp_norm(arr, 1)
```

```python
n.generalized_lp_norm(arr, 2)
```

```python
n.generalized_lp_norm(arr, 3)
```
